```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

pd.set_option('display.max_columns', None)
#pd.set_option('display.max_rows', None)
```

```python
df = pd.read_csv('telecom_customer_churn.csv')
```

```python
df.info()
```

```python
df.describe(include='all')
```

```python
df.head()
```

```python
#Count plot for Customer Status

plt.figure(figsize=(8,6))
sns.countplot(x='Customer Status', data=df, palette='Set2')
plt.title('Customer Status: Stayed, Churned or Joined')
plt.xlabel('Customer Status')
plt.ylabel('Number of Customers')
plt.show()
```

```python
churn_category_df= df[df['Churn Category'].notnull()]
```

```python
#Count plot for Churn Category

category_order = churn_category_df['Churn Category'].value_counts().index

plt.figure(figsize= (10,6))
sns.countplot(y='Churn Category', data=churn_category_df, palette='Set3',order= category_order)
plt.title('Churn Category Distribution')
plt.xlabel('Number of Customers')
plt.ylabel('Churn Category')
plt.show()
```

```python
churn_reason_df = df[df['Churn Reason'].notnull()]
```

```python
#Churn reason countplot

reason_order= churn_reason_df['Churn Reason'].value_counts().index

plt.figure(figsize= (10,6))
sns.countplot(y='Churn Reason', data= churn_reason_df, palette='Set1', order=reason_order)
plt.title('Churn reason distribution')
plt.xlabel('Number of Customers')
plt.ylabel('Churn Reason')
plt.show()
```
